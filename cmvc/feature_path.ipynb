```python
def iter_children(path):
    """
    path以下の構造をイテレーションする
    """
    for i in path.iterdir():
        if i.is_dir():
            yield from iter_children(i)
        else:
            yield i
```

```python
def mk(o_path, path):
    """
    指定したディレクトリ(o_path)以下のwavが入っている
    """
    for j in {i.parent for i in iter_children(o_path) if i.suffix == ".wav"}:
        chinpo = j.relative_to(o_path)
        print(path / o_path.name / chinpo)
```

```python
def define_property(self, name, value=None, readable=True, writable=True):
    # "_User__name" のような name mangling 後の名前.
    field_name = "_{}__{}".format(self.__class__.__name__, name)

    # 初期値を設定する.
    setattr(self, field_name, value)

    # getter/setter を生成し, プロパティを定義する.
    getter = (lambda self: getattr(self, field_name)) if readable else None
    setter = (lambda self, value: setattr(self, field_name, value)) if writable else None
    setattr(self.__class__, name, property(getter, setter))
```

```python
def define_properties(constructor=None, *, accessible=(), readable=(), writable=()):
    if callable(constructor):
        def wrapper(self, *args, **kwargs):
            for name, value in kwargs.items():
                define_property(self, name, value)
            constructor(self, *args, **kwargs)
        return wrapper
    else:
        to_set = lambda x: set(x) if any(isinstance(x, type_) for type_ in (set, list, tuple)) else {x}
        accessibles = to_set(accessible)
        readables = accessibles | to_set(readable)
        writables = accessibles | to_set(writable)

        def decorator(constructor):
            def wrapper(self, *args, **kwargs):
                for name in (readables | writables):
                    readable = name in readables
                    writable = name in writables
                    initial_value = kwargs.get(name, None)
                    define_property(self, name, initial_value, readable, writable)
                constructor_kwargs = dict([(key, kwargs[key]) for key in (constructor.__kwdefaults__ or {}) if key in kwargs])
                constructor(self, *args, **constructor_kwargs)
            return wrapper
        return decorator
```

```python
from pathlib import Path
import numpy as np

from typing import Tuple

class WaveInfo(object):
    
    Vector = "np.ndarray[np.float64]"
    
    wave_name: str = None
    person_name: str = None
    path: Path = None
    
    @define_properties(readable=("wave_name", "person_name", "path"))
    def __init__(self,
                 wave_name: str=wave_name,
                 person_name: str=person_name,
                 path: Path=path,
                ) -> None:
        pass
    
```

```python
import pyworld as pw
import pysptk as ps
import math

class Features(object):
    
    Vector = "np.ndarray[np.float64]"
    
    f0: Vector = None
    sp: Vector = None
    ap: Vector = None
    
    mc: Vector = None
    mfcc: Vector = None
    
    @define_properties
    def __init__(self,
                 f0: Vector=f0,
                 sp: Vector=sp,
                 ap: Vector=ap,
                 
                 mc: Vector=mc,
                 mfcc: Vector=mfcc,
                ) -> None:
        pass
    
    def extract_base(self,
                     ampl: Vector=None,
                     fs: int=None,
                    ) -> Tuple[Vector, Vector, Vector]:
        return pw.wav2world(ampl, fs)
    
    def extract_mc(self,
                  order: int=35,
                  fs: int=None,
                  ) -> Vector:
        alpha = round(math.log(fs)/2.9, 2)
        mc = ps.sp2mc(self.sp, order=order, alpha=alpha)
        return mc
    
    def extract_mfcc(self,
                     ampl: Vector=None,
                     fs: int=None,
                     order: int=35
                    ) -> Vector:
        return librosa.feature.mfcc(ampl, fs, n_mfcc=order)
    
    def extract_sp(self):
        pass
    
```

```python
class Wave(WaveInfo, SampledWave, Features):
    
    Vector = "np.ndarray[np.float64]"
    
    wave_name: str
    person_name: str
    path: Path
    
    fs: int = 22050
    order: int = 35
        
    ex_base: bool=True
    ex_mc: bool=False
    ex_mfcc: bool=False
    
    def __init__(self,
                 wave_name: str,
                 person_name: str,
                 path: Path,
                 data_path: Path,
                 
                 fs: int=fs,
                 order: int=order,
                 
                 ex_base: bool=ex_base,
                 ex_mc: bool=ex_mc,
                 ex_mfcc: bool=ex_mfcc,
                ) -> None:
        
        WaveInfo.__init__(self,
                          wave_name=wave_name,
                          person_name=person_name,
                          path=path,
                         )
        ampl, fs = self.load_wav(data_path/path)
        SampledWave.__init__(self, ampl=ampl, fs=fs)
        Features.__init__(self)
        
        if ex_base:
            self.extract_base()
        
        if ex_mc:
            self.extract_mc()
            
        if ex_mfcc:
            self.extract_mfcc()
            
    
    def load_wav(self,path) -> Tuple[Vector, int]:
        return super().load_wav(path=path, fs=fs)
    
    def extract_base(self) -> None:
        self.f0, self.sp, self.ap = super().extract_base(self.ampl, self.fs)
    
    def extract_mc(self) -> None:
        self.mc = super().extract_mc(self.order, self.fs)
        
    def extract_mfcc(self) -> None:
        self.mfcc = super().extract_mfcc(self.ampl, self.fs, self.order)
```
